# DataPreparation Notebook

Creates cifar10-demo dataset and splits, for use in sample notebooks for experiment management.

```python
from sbrain.dataset import DataSetImageClassification,DataSetVersion,DataSetSplit
from sbrain.dataset import DataSetStatus,JobStatus,DataSetSplitStatus,DataSetVersionStatus
from sbrain.dataset import Transformation
import numpy as np
import cv2

classes = {
                'airplane': 0,
                'automobile':1,
                'bird': 2,
                'cat': 3,
                'deer': 4,
                'dog': 5,
                'frog': 6,
                'horse': 7,
                'ship': 8,
                'truck': 9
            }



dataset_name = "cifar10-demo"

def iterator_images(data_root_path):
    import glob
    result = []
    files = glob.glob("{}/*.*".format(data_root_path))
    return iter(files)

#6985_airplane.png

def iterator_labels(data_root_path):
    import glob
    files = glob.glob("{}/*.*".format(data_root_path))
    labels = []
    classes = {
                'airplane': 0,
                'automobile':1,
                'bird': 2,
                'cat': 3,
                'deer': 4,
                'dog': 5,
                'frog': 6,
                'horse': 7,
                'ship': 8,
                'truck': 9
            }
    for f in files:
        img_name =  f.split('/')[-1:][0]
        lbl_str = img_name[img_name.index('_')+1:img_name.index('.')]
        lbl_id = classes[lbl_str]
        labels.append((img_name, lbl_id))    
    return iter(labels)


job = DataSetImageClassification(name=dataset_name).create(
    
    description="cifar 10 dataset",
    source_archive_path="shared-dir/sample-notebooks/demo-data/cifar10",
    classes=classes,
    collection_date="06-19-2018",
    image_iterator=iterator_images,
    label_iterator=iterator_labels
)

import time
from IPython.display import clear_output
job = job.get_status()
print("Dataset Job Status: {}".format(job.status))
while job.status != JobStatus.COMPLETE.value and job.status != JobStatus.FAILED.value:
    clear_output(wait=True)
    job = job.get_status()
    time.sleep(2)

ds_get = DataSetImageClassification.lookup(dataset_name)

DataSetImageClassification.search(author="admin")

ds_version = ds_get.version("v1")

version_get = DataSetVersion.lookup(dataset_name=ds_get.name,dataset_version_name="v1")
print(version_get.version)

split_job = version_get.create_data_split("cifar-10-split",[68,16,16],"split by 68-16-16")

import time
from IPython.display import clear_output
job = split_job.get_status()
print("DataSetSplit Job Status: {}".format(job.status))
while job.status != JobStatus.COMPLETE.value and job.status != JobStatus.FAILED.value:
    clear_output(wait=True)
    job = split_job.get_status()
    time.sleep(2)

version_get.search_splits(split_name='cifar-10-split')


print("\n\n Split created Successfully !!!")
```
